# Data Cleaning

### Cleaning `year` column

```python
def clean_year(x):
  if ((x == "Master's") or (x.lower() == "phd")):
    return "Graduate"
  else:
    return x

sdf_cleaned["year"] = sdf_cleaned["year"].apply(clean_year)
display(sdf_cleaned["year"].loc[ (sdf_cleaned["year"] != "Junior") & (sdf_cleaned["year"] != "Sophomore") & (sdf_cleaned["year"] != "Senior") & (sdf_cleaned["year"] != "Freshman")])

display(sdf_cleaned["year"].loc[ sdf_cleaned["year"] == None])
```

### Cleaning `feel_confidence2` column

```python
sdf_cleaned["feel_confidence2"] = sdf_cleaned["feel_confidence2"].fillna(3).astype(int)
sdf_cleaned["feel_confidence2"].head(10)
```

### Cleaning `curr_units` column

```python
def clean_curr_units(x):
  if (x > 100):
    # Cut last digit of the string
    return str(int(x))[:-1]
  return x

sdf_cleaned["curr_units"] = sdf_cleaned["curr_units"].apply(clean_curr_units).fillna(0).astype(int)
sdf_cleaned["curr_units"].head(10)
```

```python
import seaborn as sns
counts = pd.crosstab(data_df.feel_confidence2, data_df.study_group_preference)
display(counts)
joint = counts / counts.sum().sum()
sns.heatmap(joint,annot = True)
```

```python
# ENTER YOUR CODE HERE
groupPreference_counts = counts.sum(axis=0)
confidence_given_groupPreference = counts.divide(groupPreference_counts, axis=1)
display(confidence_given_groupPreference)
confidence_given_groupPreference.plot.bar(stacked=True)
```

```python
from scipy.stats import chi2_contingency
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html
chi, p, dfree, expected = chi2_contingency(counts)
print("calculated chi square value: " ,chi)
print("degree of freedom: ", dfree)
```

```python
from scipy.stats import chi2
print("for 12 degree of freedom, the chi-square value needed to reject the hypothesis at the 0.001 significance level:")
chi2.ppf(1-0.001, 12)
```
